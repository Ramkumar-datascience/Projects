```python
gender=[]
full_name = []
street = []
city = []
country = []
email = []
dob = []
age = []
for i in range(10):
    page = requests.get("https://randomuser.me/api/")
    data = page.json()["results"][0]
    gender.append(data["gender"])
    full_name.append(data["name"]["title"] +" "+ data["name"]["first"] + " "+data["name"]["last"])
    street.append(data["location"]["street"]["name"])
    city.append(data["location"]["city"])
    country.append(data["location"]["country"])
    email.append(data["email"])
    dob.append(data["dob"]["date"])
    age.append(data["dob"]["age"])
```

```python
d = {"Full Name":full_name,
"Email":email,
"Gender":gender,
"Street":street,
"City":city,
"Country":country,
"DOB":dob,
"Age":age}
```

```python
pd.DataFrame(d)
```
