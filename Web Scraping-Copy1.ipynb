# For SIngle Page

```python
names = []
prices = []
RAM = []
color = []
ROM = []
battery = []
import re

page2 = requests.get('https://www.flipkart.com/search?q=mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=2')
soup = BeautifulSoup(page2.text)

name = soup.find_all('div',class_='_3wU53n')

for i in name:
    a = i.text
    names.append(a.split('(')[0])
    
price = soup.find_all('div',class_='_1vC4OE _2rQ-NK')

for i in price:
    prices.append(i.text)
    
ram = soup.find_all('li',class_='tVe95H')

for i in ram:
    b = i.text
    c = b.split('|')[0]
    r = re.findall(r'(\d)\sGB\sRAM',c)
    if r:
        RAM.append(r)
        
for i in name:
    color.append(i.text.split('(')[1].split(',')[0])
    
for i in name:
    ROM.append(i.text.split('(')[1].split(',')[1])

bat = soup.find_all('li',class_='tVe95H')

for i in bat:
    a = i.text
    b = re.findall(r'(\d+)\smAh',a)
    if b:
        battery.append(b)
```

```python
df2 = pd.DataFrame({'Mobile Name':names,
             'Price':prices,
             'Color':color,
              'RAM':RAM,
             'ROM':ROM,
             'Battery':battery})
```

# for Multiple Pages

```python
url = 'https://www.flipkart.com/search?q=mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page='
```

```python
for i in range(2,6):
    print(url+str(i))
```

```python
names = []
prices = []
RAM = []
color = []
ROM = []
battery = []
import re
import numpy as np

for i in range(2,3):
    urls = url+str(i)
    
    page2 = requests.get(urls)
    
    soup = BeautifulSoup(page2.text)

    name = soup.find_all('div',class_='_3wU53n')

    for i in name:
        a = i.text
        names.append(a.split('(')[0])

    price = soup.find_all('div',class_='_1vC4OE _2rQ-NK')

    for i in price:
        prices.append(i.text)

    ram = soup.find_all('li',class_='tVe95H')

    for i in ram:
        b = i.text
        c = b.split('|')[0]
        #print(c)
        if c:
            #print('got')
            r = re.findall(r'(\d)\sGB\sRAM',c)
            #print(r)
            if r:
                RAM.append(r)
        else:
            RAM.append(np.nan)

    for i in name:
        if '(' in i.text:
            
            c = i.text.split('(')[1].split(',')[0]
            if len(c) == 1:
                color.append(c)
        else:
            #print('got')
            color.append(np.nan)

    for i in name:
        if '(' in i.text:
            r = i.text.split('(')[1].split(',')[1]
            #print(r)
            #ROM.append(r)
            if len(r) == 1:
                ROM.append(r)
        else:
            ROM.append(np.nan)
   

    bat = soup.find_all('li',class_='tVe95H')

    for i in bat:
        a = i.text
        b = re.findall(r'(\d+)\smAh',a)
        if b:
            battery.append(b)
        else:
            battery.append(np.nan)
            
print(len(names))
print(len(prices))
print(len(RAM))
print(len(battery))
print(len(ROM))
print(len(color))
```

# Data Cleaning

```python
pr=[]
for i in final_df['Price']:
    pr.append(re.sub(r'[₹,]','',i))
    
final_df['Price'] = pr
```

```python
rom=[]
for i in final_df['ROM']:
    #print(i)
    rom.append(re.sub(r'[A-Z\W]','',i))

final_df['ROM'] = rom
```

```python
ram = []
for i in final_df['RAM']:
    ram.append(i[0])
    
final_df['RAM']=ram
```

```python
batt = []
for i in final_df['Battery']:
    batt.append(i[0])
    
final_df['Battery']=batt
```

```python
mobile = []
for i in final_df['Mobile Name']:
    mobile.append(i.split()[0])
    
final_df['Brand'] = mobile
```
